26. Remove Duplicates from Sorted Array

Given an integer array `nums` sorted in *non-decreasing order*, remove the duplicates in-place such that each unique element appears only *once*. The *relative order* of the elements should be kept the *same*.

Since it is impossible to change the length of the array in some languages, you must instead have the result be placed in the *first part* of the array `nums`. More formally, if there are `k` elements after removing the duplicates, then the first `k` elements of nums should hold the final result. It does not matter what you leave beyond the first `k` elements.

Return `k` after placing the final result in the first `k` slots of nums.

Do *not* allocate extra space for another array. You must do this by *modifying the input array* in-place with O(1) extra memory.

**Custom Judge**:

The judge will test your solution with the following code:
```javascript
    int[] nums = [...]; // Input array
    int[] expectedNums = [...]; // The expected answer with correct length

    int k = removeDuplicates(nums); // Calls your implementation

    assert k == expectedNums.length;
    for (int i = 0; i < k; i++) {
        assert nums[i] == expectedNums[i];
    }
```

If all assertions pass, then your solution will be accepted.


**Constraints**:
   - `0 <= nums.length <= 3 * 104`
   - `-100 <= nums[i] <= 100`
   - `nums` is sorted in non-decreasing order.

````python
def judge(nums,expected,k):
    assert k == len(expected)
    for i in range(k):
        assert nums[i] == expected[i]
````

### Approach 1 (naive)
- For each element in the list
    - if duplicate, replace i.e. where count>1 with `null`
    - else, increment the total and sort it by swapping with `null` space
return the total

````python

def removeDuplicates(nums):
    """
    - Replace duplicate with None,
    - Sort by swapping the last
    """
    total=0
    for index,num in enumerate(nums):
        if nums.count(num)>1:
            nums[index] = None
        else:
            total+=1
            #find a None before current index and swap
            for i in range(index):
                if nums[i] is None:
                    nums[i]= num
                    nums[index] = None
                    break
    return total

x1=[1,1,1,1,1]
x2=[1,1,1,1,2]
x3=[1,1,2]
x4=[1,1,1,2]
x5=[0,0,1,1,1,2,2,3,3,4]

judge(x1,[1],removeDuplicates(x1))
judge(x2,[1,2],removeDuplicates(x2))
judge(x3,[1,2],removeDuplicates(x3))
judge(x4,[1,2],removeDuplicates(x4))
judge(x5,[0,1,2,3,4],removeDuplicates(x5))

#judge(x6,y6,removeDuplicates(x6))
````

## Approach 2
- set last_sorted index and total as zero
- For each element in the list:
    - if current value is none, skip below
    - else if not on last index, check any consecutive duplicate elements and replace with none
    - if current index is not equal last_sorted_index, swap the value and increment last_sorted

````python
def removeDuplicates(nums):
    sorted_last=0 #last sorted index
    total =0 #total unique items
    length =len(nums) #list length
    for index,num in enumerate(nums):
        if num is None:
            continue
        #check if consecutive items are duplicate
        if index+1 != length:
            for i in range(index+1,length):
                if nums[i] == num:
                    nums[i] = None
                else:
                    #stop
                    break
        #swap sort
        if nums[sorted_last] != num:
            sorted_last+=1
            nums[sorted_last] = num
            #nums[index]=None
            
        total+=1
    return total

x1=[1,1,1,1,1]
x2=[1,1,1,1,2]
x3=[1,1,2]
x4=[1,1,1,2]
x5=[0,0,1,1,1,2,2,3,3,4]
x7=[1,2]

judge(x1,[1],removeDuplicates(x1))
judge(x2,[1,2],removeDuplicates(x2))
judge(x3,[1,2],removeDuplicates(x3))
judge(x4,[1,2],removeDuplicates(x4))
judge(x5,[0,1,2,3,4],removeDuplicates(x5))

judge(x6,y6,removeDuplicates(x6))     

judge(x7,[1,2],removeDuplicates(x7))
````

## Approach 3
- Set total as length of the list
- Set k as 1 (will hold unique values count)
- Set last unique index as 0 
- Set index as zero (our current loop index)

- while index is less than total:
    - if current index value is not equal to last:
        - increment k
        - increment last unique index
        - move the current value to the last unique index
    - increment index

````python
def removeDuplicates(nums):
    k=1
    total=len(nums)
    index= 0
    last_unique_index = 0

    while index < total:
        if last_unique_index!=index and nums[index] != nums[last_unique_index]:
            #swap
            last_unique_index+=1
            nums[last_unique_index] = nums[index]    
            k+=1        
        index+=1
    return k

x1=[1,1,1,1,1]
x2=[1,1,1,1,2]
x3=[1,1,2]
x4=[1,1,1,2]
x5=[0,0,1,1,1,2,2,3,3,4]
x7=[1,2]

judge(x1,[1],removeDuplicates(x1))
judge(x2,[1,2],removeDuplicates(x2))
judge(x3,[1,2],removeDuplicates(x3))
judge(x4,[1,2],removeDuplicates(x4))
judge(x5,[0,1,2,3,4],removeDuplicates(x5))
judge(x7,[1,2],removeDuplicates(x7))
judge(x6,y6,removeDuplicates(x6))     
````
